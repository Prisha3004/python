```python
employees=pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_id':[10,20,10,30]
})
department=pd.DataFrame({
    'dept_id':[10,20,40],
    'dept_name':['HR','IT','FIN']
})
print(employees)
print(department)
print(pd.merge(employees,department,on='dept_id',how = 'inner'))
print(employees.merge(department,on='dept_id',how = 'inner'))
```

```python
employees=pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_a_id':[10,20,10,30]
})
department=pd.DataFrame({
    'dept_id':[10,20,40],
    'dept_name':['HR','IT','FIN']
})
print(pd.merge(employees,department,left_on = 'dept_a_id',right_on='dept_id',how = 'inner')
      .drop('dept_a_id',axis=1))
```

```python
employees=pd.DataFrame({
    'emp_id':[1,2,3,4],
    'name':['A','B','C','D'],
    'dept_id':[10,20,10,30],
    'salary':[50,30,20,70]
})
department=pd.DataFrame({
    'dept_id':[10,20,40],
    'dept_name':['HR','IT','FIN'],
    'salary':[10,70,30]
})
print(pd.merge(employees,department,on='dept_id',how = 'outer',suffixes = ('_emp','_dept'),indicator= True))
```
