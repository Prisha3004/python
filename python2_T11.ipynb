```python
import pandas as pd
import numpy as np
```

```python
df = {
    'A':[np.nan,2,3,4,np.nan],
    'B':[np.nan,5,6,7,8],
    'C':[np.nan,7,8,np.nan,9]
}
df = pd.DataFrame(df)
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(df.mean())
df
```

```python
df['A'] = df['A'].fillna(df['A'].median())
df
```

```python
df['D'] = ['V','V','H','A',np.nan]
df
```

```python
df['D'] = df['D'].fillna(df['D'].mode()[0])
df
```

```python
df['D'].mode()
```

```python
df['B'] = df['B'].fillna(7)
df
```
